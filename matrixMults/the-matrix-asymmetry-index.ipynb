```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matrix asymmetry index

n = 4

A = np.random.rand(n, n)

A = (A - A.T)/2 # formula for creating an asymmetric matrix

A_tilda = (A - A.T)/2

A_tilda_norm = np.linalg.norm(A_tilda, 'fro')

A_norm = np.linalg.norm(A,'fro')

ai = A_tilda_norm / A_norm


def MAI(M):
    M_tilda = (M - M.T)/2

    M_tilda_norm = np.linalg.norm(M_tilda)

    M_norm = np.linalg.norm(M)

    return M_tilda_norm / M_norm


```

```python
ps = np.linspace(0, 1, 100)

mai = np.zeros(len(ps))

for i in range(len(ps)):
    p = ps[i]
    A = np.random.rand(n, n)
    B = (1-p)*(A+A.T)/2 + p*(A-A.T)/2
    mai[i] = MAI(B)

plt.plot(ps, mai, 's-')
plt.show()

```
