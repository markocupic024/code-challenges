```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10

F = np.zeros((n, n), dtype='complex')
x = np.random.rand(n)

w = np.exp(-2 * np.pi * 1j / n)

for j in range(0, n):
    for k in range(0, n):
        m = j * k
        F[j, k] = w ** m

X1 = F @ x  

X2 = np.fft.fft(x)

plt.plot(np.arange(n), np.abs(X1), label='X1')
plt.plot(np.arange(n), np.abs(X2), 'o', label='X2')
plt.legend()




```
