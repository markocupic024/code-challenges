```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
thetas = np.linspace(0, 2*np.pi, 100)

XY = np.array([np.sin(thetas), np.cos(thetas)]).T

plt.figure(1, figsize=(5,5))

plt.scatter(XY[:, 0], XY[:, 1], s=80, facecolors='none', edgecolors='b')

A = np.array([[2, 1], [1, 2]])

plt.scatter(XY@A[:, 0], XY@A[:, 1], s=80, facecolors='none', edgecolors='r')
```
