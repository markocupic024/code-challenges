```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[-2,3],[2,8]])

n = 60

xi = np.linspace(-2, 2, n)

qf = np.zeros((n,n))

qfn = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        x = [xi[i], xi[j]]
        qf[i,j] = x@A@x
        qfn[i,j] = (x@A@x)/np.dot(x,x)

X, Y = np.meshgrid(xi, xi)

fig = plt.figure(figsize=(5,5))
ax1 = fig.add_subplot(111,projection='3d')
surf1 = ax1.plot_surface(X, Y, qf.T)
ax1.view_init(azim=-30, elev=30)
ax1.set_title('Raw quadratic form')

fig = plt.figure(figsize=(5,5))
ax2 = fig.add_subplot(111,projection='3d')
surf1 = ax2.plot_surface(X, Y, qfn.T)
ax2.view_init(azim=-30, elev=30)
ax2.set_title('Normalized quadratic form')

plt.show()
```
