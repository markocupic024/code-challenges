```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([ 2, 3 ])

v = np.array([ 4, 0 ])

wvp = (w.dot(v)/v.dot(v))*v

wvo = w - wvp

plt.plot([0, w[0]], [0, w[1]], 'r')
plt.plot([0, v[0]], [0, v[1]], 'b')
plt.plot([0, wvp[0]], [0, wvp[1]], 'r--')
plt.plot([0, wvo[0]], [0, wvo[1]], 'r:')

plt.legend(['w', 'v', 'w$_{||v}$', 'w$_{\perp v}$'])
plt.axis('square')
plt.grid()
plt.axis([-5, 5, -5, 5])
plt.show()
```
