```python
import numpy as np
from scipy.linalg import eig, eigh
import scipy.io
import matplotlib.pyplot as plt
```

```python
S = np.random.rand(2,2)
R = np.random.rand(2,2)

D, V = eig(S, R)
plt.xlim([-3,3]), plt.ylim([-3,3])
plt.plot([-3,3],[0,0],'k:')
plt.plot([0,0],[-3,3],'k:')
plt.plot([0, V[0,0]], [0, V[1,0]], 'k',linewidth=4)
plt.plot([0, V[0,1]], [0, V[1,1]], 'k--',linewidth=4)

D, V = np.linalg.eig(np.linalg.inv(R)@S)
plt.xlim([-3,3]), plt.ylim([-3,3])
plt.plot([-3,3],[0,0],'k:')
plt.plot([0,0],[-3,3],'k:')
plt.plot([0, V[0,0]], [0, V[1,0]], 'r',linewidth=4)
plt.plot([0, V[0,1]], [0, V[1,1]], 'r--',linewidth=4)
plt.show()

data = scipy.io.loadmat('real_matrices.mat')
S = data['S']
R = data['R']

D, V = eigh(S, R)
plt.plot(D, 's-')

D, V = eigh(np.linalg.inv(R)@S)
plt.plot(D, 's-')
plt.show()
```
