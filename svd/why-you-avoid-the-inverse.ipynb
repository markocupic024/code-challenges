```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ks = np.linspace(10, 10**12, 40)
ns = np.arange(2, 71)

R = np.zeros((len(ns), len(ks)))

for idx, n in enumerate(ns):
    U, _ = np.linalg.qr(np.random.rand(n, n))
    V, _ = np.linalg.qr(np.random.rand(n, n))
    for i in range(len(ks)):
        k = ks[i]
        S = np.diag(np.linspace(k, 1, n))
        A = U @ S @ V.T
        Ai = np.linalg.inv(A)
        I = A @ Ai
        R[idx, i] = np.abs(np.linalg.norm(I - np.eye(n)))

fig = plt.figure(figsize=(7, 7))
plt.pcolor(ks, ns, R, vmax=np.max(R) * 0.6)
plt.xlabel('Condition number')
plt.ylabel('Matrix size')
plt.colorbar()
plt.show()

    
```
