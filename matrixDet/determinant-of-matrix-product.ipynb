```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.random.rand(3,3)

B = np.random.rand(3,3)

print(np.linalg.det(A@B))

print(np.linalg.det(A) * np.linalg.det(B))

dets = np.zeros((40, 2))

for i in range(40):
    A = np.random.rand(i,i)

    B = np.random.rand(i,i)

    dets[i, 0] = np.linalg.det(A@B)
    dets[i, 1] = np.linalg.det(A) * np.linalg.det(B)

plt.plot(dets[:, 0] - dets[:, 1])
plt.show()


```
