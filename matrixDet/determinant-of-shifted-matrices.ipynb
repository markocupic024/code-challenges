```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambdas = np.linspace(0, 0.1, 30)

i = np.eye(20)

dets = np.zeros(1000)

avgDets = np.zeros(len(lambdas))

for det in range(len(lambdas)):
    lambd = lambdas[det]
    for j in range(1000):
        A = np.random.randn(20, 20)
        A[:, 0] = A[:, 1]
        As = A + lambd*i
        absDet = np.abs(np.linalg.det(As))
        dets[j] = absDet
    avgDets[det] = np.mean(dets)

plt.plot(lambdas, avgDets, 's-')
plt.xlabel('Lambda')
plt.ylabel('Average Determinant')
plt.title('Average Determinant vs. Lambda')
plt.show()
```
